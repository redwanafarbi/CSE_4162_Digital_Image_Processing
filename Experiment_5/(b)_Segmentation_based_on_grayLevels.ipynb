```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Function of Gray Level Segmentation

```python
def gray_level_segmentation(image):
    result = np.zeros_like(image)
    height, width = image.shape

    for r in range(height):
        for c in range(width):
            val = image[r][c]
            if val < 85:
                result[r][c] = 0
            elif val < 170:
                result[r][c] = 150
            else:
                result[r][c] = 255
    
    return result
```

# Load Image

```python
image = cv2.imread('cards.tif', cv2.IMREAD_GRAYSCALE)

segmented_image = gray_level_segmentation(image)

plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(segmented_image, cmap='gray')
plt.title('Gray-Level Segmentation')
plt.show()
```
