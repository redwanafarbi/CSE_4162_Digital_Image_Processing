```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Input Image

image = cv2.imread("Dollar.tif", cv2.IMREAD_GRAYSCALE)
# image = cv2.resize(image,(512,512))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("The 100 Dollar Image")
plt.show()
```

```python
# Image obtain by last 3 bits

mask = "11100000"
three_bit_image = image & int(mask,2)
plt.imshow(cv2.cvtColor(three_bit_image, cv2.COLOR_BGR2RGB))
plt.title("Image using Last 3-Bits")
plt.show()
```

```python
# Find Difference

difference_image = cv2.absdiff(np.array(image), np.array(three_bit_image))
plt.imshow(cv2.cvtColor(difference_image, cv2.COLOR_BGR2RGB))
plt.title("The Difference Image")
plt.show()
```
